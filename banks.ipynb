```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the total_assets.csv file
df = pd.read_csv('data/total_income_differ.csv', index_col=0)

# Assuming 'privatbank' is the column name for PrivatBank, plot its data
bank_name = 'privatbank'

# Check if the bank_name exists as a column
if bank_name in df.columns:
    # Plotting the data for PrivatBank
    plt.figure(figsize=(10, 6))
    plt.plot(pd.to_datetime(df.index), df[bank_name], marker='o', linestyle='-', color='b', label=bank_name.capitalize())
    plt.title(f'Total Assets for {bank_name.capitalize()}')
    plt.xlabel('Date')
    plt.ylabel('Total Assets')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print(f"Data for {bank_name.capitalize()} not found in the file.")
```
