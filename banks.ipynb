```python
def aggregate_data(column_name, output_csv):
    # Base directory containing the dataset
    base_dir = 'original_dataset/aggregation'

    # Dictionary to store data
    data_dict = {}

    # Traverse through each year folder
    for year in os.listdir(base_dir):
        year_path = os.path.join(base_dir, year)
        if os.path.isdir(year_path):
            # Traverse through each month file in the year folder
            for month_file in os.listdir(year_path):
                month_path = os.path.join(year_path, month_file)
                if month_file.endswith('.xlsx'):
                    # Extract date from the filename (assuming filename is in 'YYYY-MM.xlsx' format)
                    date_str = f"{year}-{month_file.split('-')[1]}"
                    # Read the specific column from the Excel file
                    df = pd.read_excel(month_path, skiprows=3, nrows=9, usecols=[column_name])
                    for _ in range(2):
                        df = df.drop(0).reset_index(drop=True)
                    # Assuming bank names are in the first row
                    if df.empty or column_name not in df.columns:
                        continue
                    # Store data in the dictionary
                    for bank in df[column_name].columns:
                        if bank not in data_dict:
                            data_dict[bank] = {}
                        data_dict[bank][date_str] = df[column_name][bank].iloc[0]

    # Convert the dictionary to a DataFrame
    result_df = pd.DataFrame(data_dict).sort_index()

    # Save the resulting DataFrame as a CSV file
    result_df.to_csv(output_csv)

# Example usage
aggregate_data('Загальні активи, усього', 'output.csv')
```
