```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Load total_income.csv and total_assets.csv
df_profit = pd.read_csv('data/quarterly/profit.csv', index_col=0)
df_assets = pd.read_csv('data/quarterly/total_assets.csv', index_col=0)

# Calculate the ratio of income to assets for each bank
ratio_df = (df_profit / df_assets) * 100

# Plotting
plt.figure(figsize=(14, 8))

for bank in ratio_df.columns:
    plt.plot(pd.to_datetime(ratio_df.index), ratio_df[bank], label=bank.capitalize())

plt.title('ROA')
plt.xlabel('Date')
plt.ylabel('ROA (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a custom colormap
colors = ["#FFFFD8", "#D9EDBF", "#8ECAE6", "#003049"]
n_bins = 100  # Discretizes the interpolation into bins
cmap_name = 'custom_palette'

custom_cmap = LinearSegmentedColormap.from_list(cmap_name, colors, N=n_bins)

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(ratio_df.T, cmap=custom_cmap)  # Transpose ratio_df using .T

# Customizing the plot
plt.title(' ROA   ')
plt.xlabel('')  # Now dates will be on the x-axis
plt.ylabel('')  # Banks will be on the y-axis

banks = ['', '', '', '', '', ' ']
years = ['2020', '2021', '2022', '2023', '2024']
# Adjusting ticks and labels for better visibility
ticks_every_5th = range(0, len(ratio_df.index), 4)
plt.xticks(ticks=ticks_every_5th, labels=years, rotation=0)  # Show every 5th date
plt.yticks(ticks=range(len(banks)), labels=banks, rotation=0)   # Keep y-axis labels horizontal

plt.show()
```

```python
def plot_file(name, sel_banks, path):
    df = pd.read_csv(path + name)
    df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0])
    df.set_index(df.iloc[:, 0], inplace=True)
    plt.figure(figsize=(14, 8))
    plt.style.use('seaborn-v0_8-poster')
    plt.title(os.path.splitext(name)[0])
    for bank in sel_banks:
        plt.plot(df.index, df[bank[0]], color=bank[1], label=bank[0])
    plt.legend()
    plt.grid(visible=False)
    plt.show()
    
all_banks = [['privatbank', 'lime'], ['oschadbank', 'black'], ['sense', 'blue'], ['ukreximbank','magenta'], ['ukrgasbank', 'red'], ['first investment bank', 'orange']]

big_bank = [['privatbank', 'lime'], ['oschadbank', 'black']]
medium_bank = [['sense', 'blue'], ['ukreximbank','magenta'], ['ukrgasbank', 'red']]
micro_bank = [['first investment bank', 'orange']]

for files in ['total_income.csv', 'administrative_expenses.csv']:
    plot_file(files, all_banks, 'data/quarterly/')
    plot_file(files, big_bank, 'data/quarterly/')
    plot_file(files, medium_bank, 'data/quarterly/')
    plot_file(files, micro_bank, 'data/quarterly/')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load income and expenses data
df_income = pd.read_csv('data/dollarized_quaterly/total_income.csv', index_col=0)
df_expenses = pd.read_csv('data/dollarized_quaterly/total_expenses.csv', index_col=0)

print(df_income['privatbank'])
# Assuming both dataframes have the same structure, check for consistency
if df_income.shape != df_expenses.shape or not df_income.columns.equals(df_expenses.columns):
    raise ValueError("Income and expenses dataframes do not match in shape or columns.")

# Prepare data for plotting
banks = df_income.columns
banks_ukr = ['', '', '', '', ' ', '  ']

# Plotting histograms for each bank
for index, bank in enumerate(banks):
    income = df_income[bank].values / 1000
    expenses = df_expenses[bank].values / 1000
    
    income[-1] = income[-1] / 2 * 3
    expenses[-1] = expenses[-1] / 2 * 3

    print(income[-1])
    print(expenses[-1])
    
    months = np.arange(len(income))  # Assuming each row represents a month

    bar_width = 0.4
    plt.figure(figsize=(14, 8))

    # Plotting each income bar individually to set the transparency for the last bar
    for i in range(len(income)):
        if i == len(income) - 1:
            plt.bar(months[i] - bar_width/2, income[i], width=bar_width, align='center', color='#CEAD5F', alpha=0.5)
        else:
            plt.bar(months[i] - bar_width/2, income[i], width=bar_width, label='' if i == 0 else "", align='center', color='#CEAD5F')

    # Plotting each expenses bar individually to set the transparency for the last bar
    for i in range(len(expenses)):
        if i == len(expenses) - 1:
            plt.bar(months[i] + bar_width/2, expenses[i], width=bar_width, align='center', color='#003049', alpha=0.5)
        else:
            plt.bar(months[i] + bar_width/2, expenses[i], width=bar_width, label='' if i == 0 else "", align='center', color='#003049')

    plt.xlabel('')
    plt.ylabel(' (  )')
    plt.title(f'     : {banks_ukr[index]}')

    # Customize x-axis ticks to show every half year
    plt.xticks(months[::6], df_income.index[::6], rotation=0)

    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def private_state(private_file, state_file):
    df_state = pd.read_csv(state_file, index_col=0)
    df_private = pd.read_csv(private_file, index_col=0)

    numeric_cols = df_state.select_dtypes(include=np.number).columns.tolist()
    if "privatbank" in df_state.columns:
        df_state['average_state'] = (df_state[numeric_cols].sum(axis=1) - df_state['privatbank']) / 5
    else:
        df_state['average_state'] = (df_state[numeric_cols].sum(axis=1) - df_state['']) / 5

    numeric_cols_2 = df_private.select_dtypes(include=np.number).columns.tolist()
    df_private['average_private'] = df_private[numeric_cols_2].sum(axis=1) / 13

    if "privatbank" in df_state.columns:
        df_state_private = pd.DataFrame({
        '': df_state['average_state'],
        '': df_private['average_private'],
        '': df_state['privatbank']
        })
    else:
        df_state_private = pd.DataFrame({
        '': df_state['average_state'],
        '': df_private['average_private'],
        '': df_state['']
        })
    return df_state_private

df_loans = private_state('data/quarterly/private_loans.csv', 'data/quarterly/loans.csv')
df_total_assets = private_state('data/quarterly/private_total_assets.csv', 'data/quarterly/total_assets.csv')

df_loans_assets = (df_loans / df_total_assets)


# Assuming df_loans_assets is already defined and has 'date' as the index
# Reset index to make 'date' a column
df_loans_assets.reset_index(inplace=True)
print(df_loans_assets.columns)

# Extract 'year_month' from 'date'
df_loans_assets['year_month'] = df_loans_assets['date'].apply(lambda x: x[:7])



# Plotting
plt.figure(figsize=(14, 8))

# Colors for the plot
colors = {
    '': '#003049',
    '': '#90D26D',    
    '': '#219EBC'
}

for bank in df_loans_assets.columns:
    if (bank != 'year_month' and bank != 'date'):
        plt.plot(df_loans_assets['year_month'], df_loans_assets[bank], label=bank.capitalize(), color=colors[bank])

plt.title('    ')
plt.xlabel('')
plt.ylabel(' / ')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
